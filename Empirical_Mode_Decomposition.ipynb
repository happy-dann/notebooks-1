Emipirical Mode Decompositon of the Northern Annular Mode
=========================================================

_** This is now set up to use polar cap mean geopotential height anomalies, rather than the NAM **_

This notebook aims to test whether empirical mode decomposition can be applied to the NAM (calculated from ERA-40 data), and test against the plots given to me by Scott. 

Notes:

* I am using the package [`pyhht`](https://github.com/jaidevd/pyhht), which is available on GitHub. It performs the [Hilbert-Huang](http://en.wikipedia.org/wiki/Hilbert%E2%80%93Huang_transform) transform. 

* More information about EMD, and it's application to climate data can be found here in [Coughlin & Tung (2005)](http://depts.washington.edu/amath/old_website/research/articles/Tung/journals/coughlin-tungHHT05.pdf) and [Coughlin & Tung (2004)](http://ac.els-cdn.com/S0273117704002376/1-s2.0-S0273117704002376-main.pdf?_tid=0c0d338a-ba8e-11e3-845e-00000aacb361&acdnat=1396460680_b7446f2e48c6ae5e5468311b4ec260f1)

* I have included split and displaced vortex events from Seviour et al. (2013). Splits are red stars and displacements green.


## Calculating the EMD and checking it works

First of all read in a NAM file. Initiall choosing the test period to be the same as Scott's/Mark Baldwin's plots

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
import numpy as np
import cdms2
from pyhht import EMD
import save_restore
import datetime as dt
import pandas as pd

# Read in the "NAM" file over a particular range of dates
start_date = '1958-01-01'
end_date   = '2010-12-30'
f = cdms2.open('./data/ERA_gph_polar_cap_anomaly.nc')
nam = f('zg', time=(start_date, end_date))
f.close()
p = nam.getLevel()

# Normalise the NAM by standard deviation at each pressure level
for i in range(len(p)):
    nam[:,i] = (nam[:,i])/np.std(nam[:,i])

# Read in time coordinate
t = nam.getTime()
t.units = "days since 1958-1-1 12:0:0.0" #Hack to fix times at 18:00 not 12:00 
ct = t.asComponentTime()
ct_st = [str(ct[i]) for i in range(len(ct))]
#ct_st_dates = [d.split(' ')[0] for d in ct_st]
dates = [dt.datetime.strptime(d.split(' ')[0] ,'%Y-%m-%d').date() for d in ct_st]


# Read Seviour et al. events and find where they occur in timeseries
events_all = save_restore.restore('./data/cp_mitchell_seviour_dates.pypic')['seviour_dates']
# Select only events in current time-slice
event_types = [events_all[i,0].tolist() for i in range(events_all.shape[0]) if events_all[i,1] in ct_st]
event_dates_str = [events_all[i,1].tolist() for i in range(events_all.shape[0]) if events_all[i,1] in ct_st]
event_dates = [dt.datetime.strptime(d.split(' ')[0] ,'%Y-%m-%d').date() for d in event_dates_str]
# Find the index of each event in the time-slice
event_index = [dates.index(i) for i in event_dates]
# Create a pandas dataframe with all events info
events = pd.DataFrame({'types':event_types, 'dates':event_dates, 'index':event_index})

# Print out the SSW dates in the timeslice
events
```

Now perform Empirical Mode decomposition on the nam at each pressure level.

```python
nimfs = 12 # Max number of impirical mode functions
emd = np.zeros((nam.shape[0], nam.shape[1], nimfs))

# Note: EMD takes place separately at each level.
for ip in range(len(p)):
   
    emd_lev = EMD.emd(nam[:,ip], nimfs=nimfs, extrapolation='mirror')
    
    # If fewer than nimfs found add NaNs in their place.
    while emd_lev.shape[1] < nimfs:
        nans = np.empty(nam.shape[0])
        nans[:] = np.nan
        emd_lev = np.append(emd_lev, nans[:,np.newaxis], axis=1)
        
    emd[:,ip,:] = emd_lev
     
emd = np.ma.masked_where(emd != emd, emd)
```

Define a function which plots the NAM timeseries in a Baldwin & Dunkerton-like way.

```python
def nam_plot(nam, x, p, title='', clevs=None, events=None, noaxis=False, fig=None, subplot=None, show=True):
    
    if not clevs:
        clevs = [-1.5,-1,-0.5,0.5,1,1.5]
    if noaxis:
        axes = fig.add_subplot(subplot)
    else:
        fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(15,4))
    axes.contourf(x, p[:], nam.T, clevs, cmap=plt.cm.RdBu_r, extend='both')
    axes.autoscale(False)
    if events is not None:
        if type(x[0]) != datetime.date:
            raise Exception("Cannot plot events without dates on x-axis")
        for index, row in events.iterrows():
            if row['types'] == 'displ': 
                axes.scatter(row['dates'], 10,marker=(5,1), color='orange', s=200, zorder=1)
            if row['types'] == 'split':
                axes.scatter(row['dates'], 10,marker=(5,1), color='green', s=200, zorder=1)
    plt.ylim(plt.ylim()[::-1])
    plt.ylabel('Pressure (hPa)', fontsize=14)
    plt.yscale('log') 
    plt.title(title, fontsize=14, fontweight='bold')

    if type(x[0]) == datetime.date:
        years    = mdates.YearLocator()   # every year
        yearsFmt = mdates.DateFormatter('%Y')  
        plt.gca().xaxis.set_major_locator(years)
        plt.gca().xaxis.set_major_formatter(yearsFmt)
        plt.gca().autoscale_view()
        plt.gcf().autofmt_xdate()
    if show:
        plt.show()              
```

Plot the same NAM time-slice we have read in. Orange stars represent splits, and green stars displacements.

```python
nam_plot(nam, dates, p,title='ERA-40 NAM timeseries', events=events)
```

It look's pretty close. Inevitably there will be some numerical errors. Let's now have a look at what the individual modes themselves look like:

Now let's see what each empirical mode time series looks like, taking the example of modes at 10hPa. First define a function to plot the modes one on top of each other.


```python
def plot_modes(emd, x, valid_modes, events=None):
    fig = plt.figure(figsize=(20,20))
    
    for iplot in range(len(valid_modes[0])):       
        ax = fig.add_subplot(len(valid_modes[0]),1,iplot+1)
        #ax.autoscale(False)
        ax.plot(x, emd[:,iplot])
        ax.text(0.98, 0.9, 'mode '+str(valid_modes[0][iplot]), fontsize=16, 
                transform=ax.transAxes, verticalalignment='top', horizontalalignment='right')

            
        if events is not None:
            if type(x[0]) != datetime.date:
                raise Exception("Cannot plot events without dates on x-axis")
            for index, row in events.iterrows():
                if row['types'] == 'displ': 
                    ax.scatter(row['dates'], emd[row['index'], iplot],marker=(5,1), color='orange', s=200, zorder=10)
                if row['types'] == 'split':
                    ax.scatter(row['dates'], emd[row['index'], iplot],marker=(5,1), color='green', s=200, zorder=10)
   

        #set the same range for the first 5 modes
        if iplot <= 5:
            ax.set_ylim(-3,3)
        years    = mdates.YearLocator()   # every year
        yearsFmt = mdates.DateFormatter('%Y')   
        plt.gca().xaxis.set_major_locator(years)
        plt.gca().xaxis.set_major_formatter(yearsFmt)
        plt.gca().autoscale_view()
        plt.gcf().autofmt_xdate()
           
        
    plt.show()   
```

Now plot each of the modes which actually exist (`valid_modes`). Convention seems to be that the highest frequency modes are the first.

```python
emd_10hPa = emd[:,np.where(p[:]==10.),:].squeeze()
valid_modes = np.where(emd_10hPa[0,:].mask == False)
```

## Composites of Empirical Modes around SSWs

Now to construct composites of each mose (a la Baldwin & Dunkerton 2001) around split and displacement events


```python
def calc_composite(nam, events, window_len):  

    nam_comp = {}
    numbers = {}
    for index, row in events.iterrows():  
        if (row['index'] > 60) and (row['index'] < len(dates)-61):
            try:
                nam_comp[row['types']] += nam[row['index']-window_len:row['index']+window_len+1, :] 
            except KeyError:
                nam_comp[row['types']] = nam[row['index']-window_len:row['index']+window_len+1, :]
    
    numbers = {'split':(events['types'] == 'split').sum(), 'displ':(events['types'] == 'displ').sum()}               

    nam_comp['split'] /= numbers['split']
    nam_comp['displ'] /= numbers['displ']
    
    return nam_comp, numbers
```

```python
emd_comp, numbers = calc_composite(emd, events, window_len=60)
```

## Phase relationships of the emipirical modes

The phase of a signal can be defined by the Hilbert transform:

\begin{equation}
Y(t) = \frac{1}{\pi} \mathcal{P.V.} \int_\infty^{-\infty} \frac{X(\tau)}{t-\tau} \mathrm{d}\tau
\end{equation}

The phase is then

\begin{equation}
\phi(t) = \arctan\left(\frac{Y(t)}{X(t)}\right)
\end{equation}

(Or alternatively the ratio of derivatives...?)

First calculate the phases of the 10hPa and 500hPa (note the use of `np.arctan2` rather than `np.arctan` to ensure phase goes from $-\pi$ to $\pi$ ). We can then use the `np.unwrap` function to remove discontinuities. 

```python
from scipy.signal import hilbert

# Choose empirical mode
mode = 0

x = emd_10hPa[mode] 
# If using derivative definition:
#phi_10hPa = np.arctan2(np.gradient(np.imag(hilbert(x))), np.gradient(np.real(hilbert(x))))
phi_10hPa = np.arctan2(np.imag(hilbert(x)), np.real(hilbert(x)))
y = emd_500hPa[mode] 
# If using derivative definition:
#phi_500hPa = np.arctan2(np.gradient(np.imag(hilbert(y))), np.gradient(np.real(hilbert(y))))
phi_500hPa = np.arctan2(np.imag(hilbert(y)), np.real(hilbert(y)))

# Figure to test whether this is doing a sensible thing
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.plot(x[100:500])
plt.title('10hPa Empirical Mode '+str(mode))
plt.subplot(132)
plt.plot(phi_10hPa[100:500])
plt.title('Phase')
plt.subplot(133)
plt.plot(np.unwrap(phi_10hPa)[100:500])
plt.title('Unwrapped phase')
```

Now it is possible to construct the cumulative phase difference between 10hPa and 500hPa and calculate a composite of this around all SSW events. 

```python
phi_diff = np.unwrap(phi_10hPa) - np.unwrap(phi_500hPa)

for i in range(len(event_index)):
    if event_types[i] == 'split':
        try:
            phi_diff_ssw += phi_diff[event_index[i]-90:event_index[i]+90] - phi_diff[event_index[i]]
        except NameError:
            phi_diff_ssw = phi_diff[event_index[i]-90:event_index[i]+90] - phi_diff[event_index[i]]
        
```

```python
# plot the phase difference composite
plt.plot(np.arange(-90,90),phi_diff_ssw)
plt.title('Phase difference composite, mode '+str(mode))
plt.xlabel('Lag (days)')
plt.ylabel('Normalised phase difference (radians)')
```

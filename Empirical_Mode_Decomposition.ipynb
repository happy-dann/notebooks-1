Emipirical Mode Decompositon of the NAM
=======================================

This notebook aims to test whether empirical mode decomposition can be applied to the NAM (calculated from ERA-40 data), and test against the plots given to me by Scott. 

Notes:

* I am using the package [`pyhht`](https://github.com/jaidevd/pyhht), which is available on GitHub. It performs the [Hilbert-Huang](http://en.wikipedia.org/wiki/Hilbert%E2%80%93Huang_transform) transform. 

* More information about EMD, and it's application to climate data can be found here in [Coughlin & Tung (2005)](http://depts.washington.edu/amath/old_website/research/articles/Tung/journals/coughlin-tungHHT05.pdf) and [Coughlin & Tung (2004)](http://ac.els-cdn.com/S0273117704002376/1-s2.0-S0273117704002376-main.pdf?_tid=0c0d338a-ba8e-11e3-845e-00000aacb361&acdnat=1396460680_b7446f2e48c6ae5e5468311b4ec260f1)

* I have included split and displaced vortex events from Mitchell et al. (2013). Splits are red stars and displacements green.



## Calculating the EMD and checking it works

First of all read in a NAM file. Initiall choosing the test period to be the same as Scott's/Mark Baldwin's plots

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
import numpy as np
import cdms2
from pyhht import EMD
import save_restore
import datetime as dt

# Read in the NAM file over a particular range of dates
start_date = '1970-01-02'
end_date   = '2000-01-01'
f = cdms2.open('./data/nam_era40.nc')
nam = - f('nam', time=(start_date,end_date)) #negative needed (I calculated this with the wrong sign!)
f.close()
p = nam.getLevel()

# Read in time coordinate
t = nam.getTime()
ct = t.asComponentTime()
ct_st = [str(ct[i]) for i in range(len(ct))]
dates = [dt.datetime.strptime(d.split(' ')[0] ,'%Y-%m-%d').date() for d in ct_st]


# Read Mitchell et al. events and find where they occur in timeseries
Mitchell_dates = save_restore.restore('./data/cp_mitchell_dates.pypic')['mitchell_dates']
Mitchell_dates_slice = []
for idate in Mitchell_dates[:,1]:
    if idate in ct_st:
        Mitchell_dates_slice.append(idate)

event_index = [ct_st.index(x) for x in Mitchell_dates_slice]
event_dates = [dt.datetime.strptime(d.split(' ')[0] ,'%Y-%m-%d').date() for d in Mitchell_dates_slice]
event_types = [Mitchell_dates[:,0][np.where(Mitchell_dates[:,1] == i)][0] for i in Mitchell_dates_slice]


# Print out the SSW dates in the timeslice
print event_dates
```

Now perform Empirical Mode decomposition on the nam at each pressure level.

```python
nimfs = 12
emd = np.zeros((nam.shape[0], nam.shape[1], nimfs))

# Note EMD takes place separately at each level.
for ip in range(len(p)):
    
    # Note: nimfs is the maximum number of Imperical Mode Functions,
    # may find fewer than this.
    emd_lev = EMD.emd(nam[:,ip], nimfs=nimfs)
    
    # If fewer than 5 found add NaNs in their place.
    while emd_lev.shape[1] < nimfs:
        nans = np.empty(nam.shape[0])
        nans[:] = np.nan
        emd_lev = np.append(emd_lev, nans[:,np.newaxis], axis=1)
        
    emd[:,ip,:] = emd_lev
 
emd = np.ma.masked_where(emd != emd, emd)
```

Define a function which plots the NAM timeseries in a Baldwin & Dunkerton-like way.

```python
def nam_plot(nam, p, title='', clevs=[-1.5,-1,-0.5,0.5,1,1.5], event_index=[]):
    
    fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(15,4))
    axes.contourf(dates, p[:], nam.T, clevs, cmap=plt.cm.RdBu, extend='both')
    axes.autoscale(False)
    if event_index != []:
        ypos = np.zeros(len(event_index))
        ypos[:] = 10
        for i in range(len(event_dates)):
                if event_types[i] == 'split':
                    axes.scatter(event_dates[i], 10,marker=(5,1), color='red', s=150, zorder=1)
                if event_types[i] == 'displ':
                    axes.scatter(event_dates[i], 10,marker=(5,1), color='green', s=150, zorder=1)
    plt.ylim(plt.ylim()[::-1])
    plt.ylabel('Pressure (hPa)', fontsize=14)

    plt.yscale('log')
 
    plt.title(title, fontsize=14, fontweight='bold')

    years    = mdates.YearLocator()   # every year
    months   = mdates.MonthLocator()  # every month
    yearsFmt = mdates.DateFormatter('%Y')
    
    
    plt.gca().xaxis.set_major_locator(years)
    plt.gca().xaxis.set_major_formatter(yearsFmt)
    plt.gca().xaxis.set_minor_locator(months)
    plt.gca().autoscale_view()
    plt.gcf().autofmt_xdate()
    plt.show()  
    
    
    
```

Plot the same NAM timeslice as Baldwin & Dunkerton

```python
nam_plot(nam,p,title='ERA-40 NAM timeseries', event_index=event_index)
```

It should be possible to reconstruct this from the sum of all the empirical modes. Looks pretty close!

```python
nam_plot(np.ma.sum(emd, axis=2), p, title='Reconstructed from all IMFs', event_index=event_index)
```

Now, what does the first Empirical mode look like? Note the smaller countour interval than previous plots.

```python
nam_plot(emd[:,:,0], p, title='First Empirical Mode', clevs=np.arange(-1,1,0.1), event_index=event_index)
```

The modes certainly look quite barotropic, but the plot is a bit different to Scott's - we should probably check we're using the same method.

Now let's see what each empirical mode time series looks like, taking the example of modes at 10hPa. First define a function to plot the modes one on top of each other.


```python
def plot_modes(emd, valid_modes):
    plt.figure(figsize=(20,20))
    for iplot in range(len(valid_modes[0])):       
        plt.subplot(len(valid_modes[0]),1,iplot+1)
        plt.plot(dates, emd[:,iplot], label='mode '+str(valid_modes[0][iplot]))
        for i in range(len(event_dates)):
                if event_types[i] == 'split':
                    plt.scatter(event_dates[i], emd[event_index[i], iplot], marker=(5,1), color='red', s=80)

                if event_types[i] == 'displ':
                    plt.scatter(event_dates[i], emd[event_index[i], iplot], marker=(5,1), color='green', s=80)

        #plt.xlim(0,emd.shape[0])
        plt.ylim(-3,3)
        years    = mdates.YearLocator()   # every year
        months   = mdates.MonthLocator()  # every month
        yearsFmt = mdates.DateFormatter('%Y')   
        plt.gca().xaxis.set_major_locator(years)
        plt.gca().xaxis.set_major_formatter(yearsFmt)
        plt.gca().xaxis.set_minor_locator(months)
        plt.gca().autoscale_view()
        plt.gcf().autofmt_xdate()
        
        plt.legend()
    plt.show()   
```

Now plot each of the modes which actually exist (`valid_modes`). Convention seems to be that the highest frequency modes are the first.

```python
emd_10hPa = emd[:,np.where(p[:]==10.),:].squeeze()
valid_modes = np.where(emd_10hPa[0,:].mask == False)
```

```python
plot_modes(emd_10hPa, valid_modes)
```

Mode 2 seems to be the lagest amplitude, and and also peaks around the time of the two weak vortex events. Let's plot mode 2 at all levels... (note: now back to same contour interval as for the NAM).

## Covariance of tropospheric and stratospheric modes 

I am using rolling covariance with a 60-day moving window (this can be found in `pandas`) of the modes between say 10 and 500 hPa.

```python
from pandas import DataFrame
import pandas as pd
dates = pd.date_range(start_date, end_date, freq='d')
```

```python
emd_10hPa = DataFrame(emd[:,np.where(p[:]==10.),:].squeeze(), index=dates[0:nam.shape[0]])
emd_500hPa = DataFrame(emd[:,np.where(p[:]==500.),:].squeeze(), index=dates[0:nam.shape[0]])
```

```python
a = pd.stats.moments.rolling_cov(emd_10hPa[1], emd_500hPa[1], window=60)
```

```python
plt.figure(figsize=(15,4))
plt.plot(dates[0:len(a)], a)
plt.title('Coherence of 10 and 500hPa modes')
for i in range(len(event_dates)):
    if event_types[i] == 'split':
        plt.scatter(event_dates[i], a[event_index[i]],marker=(5,1), color='red', s=80, zorder=10)
    if event_types[i] == 'displ':
        plt.scatter(event_dates[i], a[event_index[i]],marker=(5,1), color='green', s=80, zorder=10)
years    = mdates.YearLocator()   # every year
months   = mdates.MonthLocator()  # every month
yearsFmt = mdates.DateFormatter('%Y')   
plt.gca().xaxis.set_major_locator(years)
plt.gca().xaxis.set_major_formatter(yearsFmt)
plt.gca().xaxis.set_minor_locator(months)
plt.gca().autoscale_view()
plt.gcf().autofmt_xdate()
```

## Observations

It looks as though the coherence (almost) always increases following a split event, but can sometimes fall following a displacement.

Emipirical Mode Decompositon of the NAM
=======================================

This notebook aims to test whether empirical mode decomposition can be applied to the NAM (calculated from ERA-40 data), and test against the plots given to me by Scott. 

Notes:

* I am using the package [`pyhht`](https://github.com/jaidevd/pyhht), which is available on GitHub. It performs the [Hilbert-Huang](http://en.wikipedia.org/wiki/Hilbert%E2%80%93Huang_transform) transform. 

* More information about EMD, and it's application to climate data can be found here in [Coughlin & Tung](http://depts.washington.edu/amath/old_website/research/articles/Tung/journals/coughlin-tungHHT05.pdf)


First of all read in a NAM file. Initiall choosing the test period to be the same as Scott's/Mark Baldwin's plots

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import cdms2
from pyhht import EMD


# Read in the NAM file over a particular range of dates
f = cdms2.open('./data/nam_era40.nc')
nam = f('nam', time=('1998-10-01','1999-05-01'))
f.close()
p = nam.getLevel()
```

Now perform Empirical Mode decomposition on the nam at each pressure level.

```python
nimfs = 12
emd = np.zeros((nam.shape[0], nam.shape[1], nimfs))

# Note EMD takes place separately at each level.
for ip in range(len(p)):
    
    # Note: nimfs is the maximum number of Imperical Mode Functions,
    # may find fewer than this.
    emd_lev = EMD.emd(nam[:,ip], nimfs=nimfs)
    
    # If fewer than 5 found add NaNs in their place.
    while emd_lev.shape[1] < nimfs:
        nans = np.empty(nam.shape[0])
        nans[:] = np.nan
        emd_lev = np.append(emd_lev, nans[:,np.newaxis], axis=1)
        
    emd[:,ip,:] = emd_lev
 
emd = np.ma.masked_where(emd != emd, emd)
```

Define a function which plots the NAM timeseries in a Baldwin & Dunkerton-like way.

```python
def nam_plot(nam, p, title='',clevs=[-1.5,-1,-0.5,0.5,1,1.5]):
    
    plt.figure(figsize=(10,4))  
    corr = plt.contourf(np.arange(nam.shape[0]), p[:], nam.T, clevs, cmap=plt.cm.RdBu_r, extend='both')
    plt.ylim(plt.ylim()[::-1])
    plt.ylabel('Pressure (hPa)', fontsize=14)
    plt.xlabel('days')
    plt.yscale('log')
    plt.xticks([0,31,61,92,123,151,182,211], 
               ('1 Oct', '1 Nov', '1 Dec', '1 Jan', '1 Feb', '1 Mar', '1 Apr','1 May')) 
    plt.title(title, fontsize=14, fontweight='bold')
    plt.show()  
    
    
    
```

Plot the same NAM timeslice as Baldwin & Dunkerton

```python
nam_plot(nam,p,title='ERA-40 NAM timeseries')
```

It should be possible to reconstruct this from the sum of all the empirical modes. Looks pretty close!

```python
nam_plot(np.ma.sum(emd, axis=2), p, title='Reconstructed from all IMFs')
```

Now, what does the first Empirical mode look like? Note the smaller countour interval than previous plots.

```python
nam_plot(emd[:,:,0], p, title='First Empirical Mode', clevs=np.arange(-1,1,0.1))
```

The modes certainly look quite barotropic, but the plot is a bit different to Scott's - we should probably check we're using the same method.

Now let's see what each empirical mode time series looks like, taking the example of modes at 10hPa. First define a function to plot the modes one on top of each other.


```python
def plot_modes(emd, valid_modes):
    plt.figure(figsize=(8,10))
    for iplot in range(len(valid_modes[0])):       
        plt.subplot(len(valid_modes[0]),1,iplot+1)
        plt.plot(emd[:,iplot], label='mode '+str(valid_modes[0][iplot]))
        plt.xlim(0,emd.shape[0])
        plt.ylim(-3,3)
        plt.legend()
    plt.show()   
```

Now plot each of the modes which actually exist (`valid_modes`). Convention seems to be that the highest frequency modes are the first.

```python
emd_10hPa = emd[:,np.where(p[:]==10.),:].squeeze()
valid_modes = np.where(emd_10hPa[0,:].mask == False)
```

```python
plot_modes(emd_10hPa, valid_modes)
```

Mode 2 seems to be the lagest amplitude, and and also peaks around the time of the two weak vortex events. Let's plot mode 2 at all levels...

```python
nam_plot(emd[:,:,2], p, title='Second Empirical Mode', clevs=np.arange(-1,1,0.1))
```

Interesting! The first event is more baroclinic than the second (is this the displacement?).

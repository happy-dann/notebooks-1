# Significance tests for the splits/displacements NAM composite

Following many suggestions to do this, I have finally got around to attempting a significance test for the difference of the NAM composites for splits and displacements. Here, my null hypothesis is that **there is no difference (in terms of the NAM) between splits and displacements**. This is different from the significance test in Mitchell et al. 2013, who test the two null hypostheses: **there is no difference (in terms of the NAM) between splits and random winter periods** and **there is no difference (in terms of the NAM) between displacements and random winter periods**. 

I will test the difference between the composites in two ways:

1. Is there a difference in the NAM averaged over some lag (default: 0-30 days) at some height (default: 1000 hPa)?
2. Is there a difference at each lag and height (we can then overlay this onto a dripping paint plot)?

Both tests proceed by a Monte-Carlo method. I randomly reassign the label 'split' or displacement' to each event, and recalculate the  composite difference (so I am sampling **without replacement**). Repeat this many times to give a distribution of composite differences and then find the percentile where the true composite difference falls. I use a two-tailed test throughout.  


---
## Time-averaged NAM

First of all, read in some data which has the NAM from -90-+90 days at all levels, for each split and displacement event.  

```python
import save_restore
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats

# restore dictionary which includes NAM for all events in Seviour et al. 2013
nam_comp = save_restore.restore('./data/nam_comp_data.pypic')['nam_comp']

# pick a level to look at
lev = 1000
ip = np.where(nam_comp['p'][:] == lev)[0]

# put all composites together
nam_comp_all = np.append(nam_comp['split_comp'], nam_comp['displ_comp'], axis=2)

# pick a range of days to look over
day_start = 90
day_end = 120

nam_comp_all = np.mean(nam_comp_all[day_start:day_end,ip,:], axis=0).squeeze()
split_comp = np.mean(nam_comp['split_comp'][day_start:day_end,ip,:], axis=0).squeeze()
displ_comp = np.mean(nam_comp['displ_comp'][day_start:day_end,ip,:], axis=0).squeeze()
```

Now define a function which will give the distribution of randomly generated composite differences

```python
def mc_composites(events, ncomp1, ncomp2, niter=1000):
    """Takes two samples of events and returns a distribution of differences 
    of composites size ncomp1, ncomp2, sampled from the events niter times."""
    
    nevents = len(events)
    ievents = np.arange(nevents)
    
    diff = np.empty(0)
    for iiter in xrange(niter):
        comp1_events = np.random.choice(ievents, ncomp1, replace=False) # No replacement??
        comp2_events = np.delete(ievents, comp1_events)
        
        comp1 = events[comp1_events]
        comp2 = events[comp2_events]
        
        diff = np.append(diff, np.mean(comp1)-np.mean(comp2))
    
    return diff     
    
```

```python
diffs = mc_composites(nam_comp_all, len(split_comp), len(displ_comp), niter=10000)
```

Plot the distribution of random composite differences alongside the observed one.

```python
plt.hist(diffs, bins=50, histtype='step', label='Random differences')
plt.vlines(np.mean(split_comp)-np.mean(displ_comp),0,500, lw=5, color='red',label='True value')
plt.title('Composite difference NAM 0-30 days following events')
plt.legend(loc='best')
```

How significant is this result?

```python
perc = stats.percentileofscore(diffs, np.mean(split_comp)-np.mean(displ_comp))
sig = 100 - 2*(100-perc) # two-tailed test
print "Significance of composite difference at %s hPa between %s-%s days: %s %%" % (lev, day_start-90, day_end-90, sig)
```

So the composite difference is not significant, and the null hypothesis can't be rejected.

## NAM at each time lag

Now we calculate the instataneous significance by random samples at each time lag and each pressure level. The following function returns the two-tailed significance for the significance at a particular lag and height.

```python
def sig_at_point(ilag, ilev, niter=1000):
    """Returns the siginicance values for a two-tailed test 
    of a difference at a particular point in lag-height space"""
    
    all_comp = nam_comp_all
    split_comp = nam_comp['split_comp'][ilag,ilev,:].squeeze()
    displ_comp = nam_comp['displ_comp'][ilag,ilev,:].squeeze()
    
    diffs = mc_composites(all_comp, len(split_comp), len(displ_comp), niter=niter)
    sig = stats.percentileofscore(diffs, np.mean(split_comp)-np.mean(displ_comp))
    if sig < 50:
        sig = 100-sig
    sig = 100 - 2*(100-sig) # for a two-tailed test 
    
    return sig            
        
```

Now actually calculate significance at each point (this takes quite a while to run with 1000 samples at each point!)

```python
sig = np.empty(nam_comp['displ_comp'].shape[0:2])

for ilev in range(len(nam_comp['p'])):
    for ilag in range(nam_comp['split_comp'].shape[0]):
                      
        sig[ilag,ilev] = sig_at_point(ilag, ilev, niter=1000) 
```

Make a function for plotting the nam with stippling (crosses for >90%, dots for >95%):

```python
def nam_plot(nam, p, title='', clevs=[-1,-0.75,-0.5,-0.25,0.25,0.5,0.75,1], sig=None):
    
    fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10,3))
    axes.contourf(np.arange(-90,90), p[:], nam.T, clevs, cmap=plt.cm.RdBu_r, extend='both')
    axes.autoscale(False)
   
    axes.set_xticks([-90,-60,-30,0,30,60,90])
    axes.set_ylim(plt.ylim()[::-1])
    axes.set_ylabel('Pressure (hPa)', fontsize=14)
    axes.set_yscale('log')
 
    if sig is not None:
        axes.contourf(np.arange(-90,90), p[:], sig.T, [95,100], colors='none', extend='neither', hatches=['.',None])
        axes.contourf(np.arange(-90,90), p[:], sig.T, [90,95], colors='none', extend='neither', hatches=['x',None])
    
    plt.title(title, fontsize=14, fontweight='bold')

    plt.show() 
```

```python
nam_plot(np.mean(nam_comp['split_comp'],axis=2), nam_comp['p'], title='Splits')
nam_plot(np.mean(nam_comp['displ_comp'],axis=2), nam_comp['p'], title='Displacements')
nam_plot(np.mean(nam_comp['split_comp'],axis=2)-np.mean(nam_comp['displ_comp'],axis=2), nam_comp['p'], title='Difference', sig=sig)
```

So it seems there are some >95% significant regions, particularly around lag 0, where splits look more barotropic than displacements. There is also a region with some significance from ~lag 0-30 in the troposphere. 

However, we should probably be a bit careful interpreting this. For instance, it seems unlikely that the significant difference at lag ~-75 is really physically meaningful.

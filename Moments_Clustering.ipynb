Clustering methods applied to the moment diagnostics
====================================================

The aim of this notebook is to test the sensitivity of clustering algorithms applied to the moment diagnostics, using `k-means` and `heirarchical` clustering from `scikit-learn`

* I have newly included the `mean-shift` algorithm: http://sociograph.blogspot.co.uk/2011/11/accessible-introduction-to-mean-shift.html


```python
import save_restore
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cdms2
import sklearn.cluster as cluster

months = [12,1,2,3]

f1 = cdms2.open('./data/era_40_int_daily_pv_aspect_noleaps.nc')
aspect = f1('aspect')
f1.close()
f2 = cdms2.open('./data/era_40_int_daily_pv_latcent_noleaps.nc')
latcent = f2('latcent')
f2.close()

t  = latcent.getTime()
ct = t.asComponentTime()

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecessary plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

```python
def mask_values(aspect, latcent, ct):
    # Mask values where either variable is NaN or inf
    index = np.zeros(aspect.shape[0])
    for i in range(aspect.shape[0]):
            if (aspect[i] == aspect[i]) and (latcent[i] == latcent[i]) and not np.isinf(aspect[i]) and not np.isinf(latcent[i]):
                    index[i] = 1 # set to 1 for good values, 0 for bad values
            if ct[i].month not in months:
                index[i] = 0
            if aspect[i] > 20:
                index[i] = 0 
    aspect_ma = np.ma.masked_where(index==0, aspect)
    latcent_ma = np.ma.masked_where(index==0, latcent)

    x = aspect_ma.compressed()
    y = latcent_ma.compressed()

    return x, y
```

```python
a, l = mask_values(aspect, latcent, ct)

a_norm = (a - np.mean(a))/np.std(a)
l_norm = (l - np.mean(l))/np.std(l)
X = np.append(a_norm[:,np.newaxis],l_norm[:,np.newaxis], axis=1)
X.shape
```

```python
plt.scatter(X[:,0], X[:,1], alpha=0.08)
```

```python
bandwidth = cluster.estimate_bandwidth(X, quantile=0.2, n_samples=1000)

estimators = {'k-means': cluster.KMeans(n_clusters=3), 
              'Hierarchical':cluster.Ward(n_clusters=3),
              'Mean-Shift': cluster.MeanShift(bandwidth=bandwidth, bin_seeding=True)}
```

```python
labels={}

for name, est in estimators.iteritems():
    
    est.fit(X)
    labels[name] = est.labels_
    
```

```python
import brewer2mpl
set2 = brewer2mpl.get_map('Dark2', 'qualitative', 5).mpl_colors
```

```python
fig = plt.figure(figsize=(5,5))
fignum = 1

for name, est in estimators.iteritems():   
    
    ax = fig.add_subplot(2,2,fignum)
    colors = [set2[i] for i in labels[name]]
    ax.scatter((X[:, 0]*np.std(a))+np.mean(a), (X[:, 1]*np.std(l))+np.mean(l), c=colors, alpha=0.2, edgecolors='none')
    ax.set_xlim(1,6)
    ax.set_ylim(50,90)
    letters = ['a','b','c']
    ax.set_title('('+letters[fignum-1]+') '+name, fontweight='bold')
    if fignum == 3:
        ax.set_xlabel('Aspect ratio', fontsize=12)
    if fignum ==1 or fignum == 3:
        ax.set_ylabel('Centroid latitude', fontsize=12)
        
        
        
        
    remove_border()
    fignum += 1

ax = fig.add_subplot(2,2,fignum)    
A = (X[:, 0]*np.std(a))+np.mean(a)
L = (X[:, 1]*np.std(l))+np.mean(l)
labels_new = np.zeros((L.shape[0]))
labels_new[np.where(A>2.4)] = 1
labels_new[np.where(L<66)] = 2
q1 = L<66
q2 = A>2.4
labels_new[np.where(q1 & q2)] = 3

colors = [set2[i] for i in labels_new.astype(np.int)]
ax.scatter(A, L, c=colors, alpha=0.2, edgecolors='none')
ax.set_xlim(1,6)
ax.set_ylim(50,90)
ax.set_xlabel('Aspect ratio', fontsize=12)
remove_border()
ax.set_title('(d) Thesholds', fontweight='bold')
plt.tight_layout()
plt.savefig('clustering.pdf')
```
